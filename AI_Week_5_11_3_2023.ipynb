ĐK Quạt làm mát

```
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
from skfuzzy import control as ctal

temperature = ctal.Antecedent(np.arange(18,28,0.1),'temperature')
sensor = ctal.Antecedent(np.arange(0,41,0.1),'sensor')
Hz = ctal.Consequent(np.arange(0,61,0.1),'Hz')

temperature['Low']=fuzz.trimf(temperature.universe,[18,18,22.5]) 
temperature['Normal']=fuzz.trimf(temperature.universe,[22,24,26.5]) 
temperature['High']=fuzz.trimf(temperature.universe,[26,28,28]) 
#################################################################################
sensor['Low']=fuzz.trimf(sensor.universe,[0,0,15]) 
sensor['Normal']=fuzz.trimf(sensor.universe,[13,20,25]) 
sensor['High']=fuzz.trimf(sensor.universe,[24,40,40]) 
#################################################################################
Hz['Low']=fuzz.trimf(Hz.universe,[0,0,21]) 
Hz['Normal']=fuzz.trimf(Hz.universe,[19,30,41]) 
Hz['High']=fuzz.trimf(Hz.universe,[39,60,60]) 

rule_1=ctal.Rule(temperature['Low']&sensor['Low'],Hz['Low'])
rule_2=ctal.Rule(temperature['Low']&sensor['Normal'],Hz['Normal'])
rule_3=ctal.Rule(temperature['Low']&sensor['High'],Hz['High'])

rule_4=ctal.Rule(temperature['Normal']&sensor['Low'],Hz['Low'])
rule_5=ctal.Rule(temperature['Normal']&sensor['Normal'],Hz['Low'])
rule_6=ctal.Rule(temperature['Normal']&sensor['High'],Hz['Normal'])

rule_7=ctal.Rule(temperature['High']&sensor['Low'],Hz['Low'])
rule_8=ctal.Rule(temperature['High']&sensor['Normal'],Hz['Low'])
rule_9=ctal.Rule(temperature['High']&sensor['High'],Hz['Low'])

AC_ctal=ctal.ControlSystem([rule_1,rule_2,rule_3,
                            rule_4,rule_5,rule_6,
                            rule_7,rule_8,rule_9])

AC=ctal.ControlSystemSimulation(AC_ctal)

AC.input['temperature']= 18
AC.input['sensor'] = 35

AC.compute()

print("AC result:\n",AC.output['Hz'])
Hz.view(sim=AC)
```
